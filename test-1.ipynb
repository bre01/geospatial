```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets._samples_generator import make_blobs
x,y=make_blobs(n_samples=15,centers=3,random_state=1,cluster_std=0.6)
plt.figure(figsize=(5,5))
plt.scatter(x[:,0],x[:,1],c=y)
##x[:,0]第一个特征的所有取值---x
##x[:,1]第二个特征的所有取值---y
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3,random_state=1)
kmeans.fit(x)
y_kmeans=kmeans.predict(x)  #聚类类别结果
centroids=kmeans.cluster_centers_  #聚类中心点
print('聚类中心点为：',centroids)

#作图
plt.scatter(x[:,0],x[:,1],c=y_kmeans,cmap='Dark2',s=50,alpha=0.7,marker='x')
plt.scatter(centroids[:,0],centroids[:,1],c=[0,1,2],cmap='Dark2',s=70,marker='o')
plt.title('K-means 300times points\n')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# 使用KMeans进行训练
kmeans = KMeans(n_clusters=3, random_state=1).fit(x)
# 获取KMeans的预测结果
y_kmeans = kmeans.predict(x)
# 创建一个新的图形
plt.figure(figsize=(10,5))
# 在右边的子图中展示KMeans的结果
plt.subplot(122)
plt.scatter(x[:, 0], x[:, 1], c=y_kmeans)
plt.title("KMeans")
# 展示图形
plt.show()
centroids=kmeans.cluster_centers_  #聚类中心点
print('聚类中心点为：',centroids)
```
